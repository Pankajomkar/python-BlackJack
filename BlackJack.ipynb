```python
# IMPORT STATEMENTS AND VARIABLE DECLARATIONS:

import random

suits = ('Hearts','Diamonds','Spades','Clubs')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Queen','King','Jack','Ace')
values = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Queen':10,
          'King':10,'Jack':10,'Ace':11}
playing = True

# CLASS DEFINTIONS:

# Consider making a Card class where each Card object has a suit and a rank,
# then a Deck class to hold all 52 Card objects, and can be shuffled, 
# and finally a Hand class that holds those Cards that have been dealt to each player from the Deck.

class Card():
    
    def __init__(self,suit,rank):
        self.rank = rank
        self.suit = suit
                   
    def __str__(self):
        return self.rank+ " of " + self.suit


# Deck class instance create all cards as soon as we create its instance 
    
class Deck():
    
    def __init__(self):
        self.deck = [];  # start with an empty list
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))               
               
    # call card class string method to get the card details 
    
    def __str__(self):
        deck_comp = ''  # start with an empty string
        for card in self.deck:
            deck_comp += '\n'+ card.__str__() # add each Card object's print string
        return "The deck has " + deck_comp 
    
    # Call a shuffle function to shuffle all the cards randomly
        
    def shuffle(self):
        random.shuffle(self.deck)       
        
    # To Pick a single card from deck we use this method
        
    def deal(self):
        single_card = self.deck.pop()
        return single_card

    
class Hand():
    
    def __init__(self):
        self.value = 0  # start with zero value
        self.cards = [] # start with an empty list as we did in the Deck class
        self.aces = 0  # add an attribute to keep track of aces
        
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank]        
        if card.rank =='Ace':
            self.aces += 1  # add to self.aces
            
    def adjust_for_ace(self):         
        while self.value > 21 and self.aces > 0 :
            self.value -= 10
            self.aces -= 1       

class Chips():
    
    def __init__(self,total=100):
        self.total = total
        self.bet = 0
        print("Your valet has {}".format(self.total))
        
    def win_bet(self):
        self.total += self.bet
        
    def loose_bet(self):
        self.total -= self.bet
        
# FUNCTION DEFINITIONS:

def take_bet(chips):
    
    while True:        
        try :
            chips.bet = int(input("How many chips would you like to bet ? "))
        except ValueError:
            print("Sorry Please provide a valid input")
        else:
            if chips.bet > chips.total :
                print('Sorry you don\'t have enough chips! You have {} '.format(chips.total))
            else:
                break
        
def hit(deck,hand):
    
    single_card=deck.deal()
    hand.add_card(single_card)
    hand.adjust_for_ace()       
    
    
def hit_or_stand(deck,hand):
    global playing    
    while True:
        try:
            print("\n--------------------\n")
            x=int(input("Hit or Stand ?"+ "\n 1) Hit \n 2) Stand \n Enter your choice: "))
        except ValueError:
            print("Please enter a valid option")
        else:
            print("Your choice is {}".format(x))
            print("\n--------------------\n")
            if x == 1:
                hit(deck,hand)
            elif x == 2:
                print("Player Stands Dealer's Turn")
                playing = False
            else:
                print("Please Enter valid Option 1 or 2")
                continue
            break        
        
def show_some(player,dealer):
    
    print("--------------------")
    print('Dealers Hand')
    print(' <card hidden>')
    print('',dealer.cards[1])
    print("\nPlayer cards:", *player.cards, sep='\n ')
    print("Player value =",player.value)

def show_all(player,dealer):
    
    print("--------------------")
    print('Dealers Hand')
    #for card in dealer.cards:
    #    print(card)
    print("\nDealer cards:", *dealer.cards, sep='\n ')
    print("Dealer value =",dealer.value)
    #for card in player.cards:
    #    print(card)
    print("\nPlayer cards:", *player.cards, sep='\n ')
    print("Player value =",player.value)
    
    
def player_busts(chips):
    print("Player Bust")
    chips.loose_bet()
    
def player_wins(chips):
    print("Player Wins")
    chips.win_bet()
    
def dealer_busts(chips):
    print("Player Wins Dealer Busted")
    chips.win_bet()
    
def dealer_wins(chips):
    print("Player Bust Dealer Wins")
    chips.loose_bet()
    
def push():
    print("Dealer and Player Tie! PUSH")
    
def request_chips():
    while True:
        try:
            chips_valet = int(input("Enter the amount to be added in your valet "))
        except ValueError:
            print("please enter a valid currenncy")
        else:
            break
    return chips_valet        

# GAMEPLAY!

while True:    
    print('\n\nWelcome to BlackJack! Get as close to 21 as you can without going over!\n\
    Dealer hits until he/she reaches 17. Aces count as 1 or 11.')
    
    # Create & shuffle the deck, deal two cards to each player
    deck = Deck()
    deck.shuffle()   
    
    # Player now have 2 cards
    player_hand = Hand()
    player_hand.add_card(deck.deal())    
    player_hand.add_card(deck.deal())
    
    # Dealer now have 2 cards
    dealer_hand = Hand()
    dealer_hand.add_card(deck.deal())    
    dealer_hand.add_card(deck.deal())
    
    # Set up the Player's chips
    chips_valet = request_chips()   
    player_chips = Chips(chips_valet)   
    # remember the default value is 100
    
    # Prompt the Player for their bet:
    take_bet(player_chips)
    
    # Show the cards:
    show_some(player_hand,dealer_hand)
    
    while playing: # recall this variable from our hit_or_stand function
        
        # Prompt for Player to Hit or Stand
        hit_or_stand(deck,player_hand)        
        show_some(player_hand,dealer_hand)
        
        if player_hand.value > 21:
            player_busts(player_chips)
            break       
    
    # If Player hasn't busted, play Dealer's hand
    if player_hand.value <= 21:
        
        while dealer_hand.value < 17:
            hit(deck,dealer_hand)     
    
        # Show all cards
        show_all(player_hand,dealer_hand)

         # Test different winning scenarios
        if dealer_hand.value > 21:
            dealer_busts(player_chips)
            
        elif dealer_hand.value > player_hand.value:
            dealer_wins(player_chips)
            
        elif player_hand.value > dealer_hand.value:
            player_wins(player_chips)
            
        else:
            push()
            
     # Inform Player of their chips total       
    print("\n Player total chips = {}".format(player_chips.total))

    # Ask to play again
    new_game = input("Would you like to play new game? y/n")
    if new_game[0].lower() == 'y':
        playing = True
        continue
    else:
        print("Thank you for playing ")
        break
```
